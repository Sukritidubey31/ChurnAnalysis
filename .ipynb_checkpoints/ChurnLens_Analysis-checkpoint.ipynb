## Section 1 — Data Overview & First Look

### What is TheLook?

TheLook is a **synthetic but realistic ecommerce dataset** maintained by the Google Looker team and hosted as a public BigQuery dataset. It was chosen over commonly-used datasets (Telco churn, UCI bank churn) for three reasons:

1. **Multi-table structure** — requires real SQL joins across `users`, `orders`, `order_items`, and `events` tables, mirroring actual production data engineering
2. **US-based customers** — relevant for domestic roles and US market analysis
3. **Behavioral richness** — includes web events, traffic sources, and product categories alongside transactional data
4. **No pre-labeled churn column** — requires defining your own churn metric, which demonstrates analytical judgment

### Churn Definition

Since TheLook has no built-in churn label, we defined churn as:

> **A customer is churned if their last completed order was more than 90 days before the dataset's reference date (max order date)**

This is a standard industry definition for ecommerce churn and was implemented in BigQuery SQL before any modeling began.

```python
df_raw = pd.read_csv(DATA_PATH)
print(f'Shape: {df_raw.shape}')
print(f'\nColumns:\n{list(df_raw.columns)}')
print(f'\nData types:\n{df_raw.dtypes}')
```

```python
print('=== Basic Stats ===')
print(f"Total customers: {len(df_raw):,}")
print(f"Churn rate (all users): {df_raw['churned'].mean():.1%}")
print(f"\nMissing values:")
print(df_raw.isnull().sum()[df_raw.isnull().sum() > 0])
print(f"\nNumerical summary:")
df_raw.describe().round(2)
```

```python
# Scope to buyers only — users with at least 1 completed order
# Non-buyers are trivially churned and don't help the model learn behavioral signals
df = df_raw[df_raw['total_orders'] >= 1].copy()

print(f'All users: {len(df_raw):,} | Churn rate: {df_raw["churned"].mean():.1%}')
print(f'Buyers only: {len(df):,} | Churn rate: {df["churned"].mean():.1%}')
print(f'\nScoping to buyers reduces noise and focuses the model on behavioral patterns')
print(f'that actually distinguish retained vs churned customers.')
```

## Section 3 — Feature Engineering Decisions

Every feature was created to answer a specific business question. Here's the reasoning behind each one.

```python
# Feature engineering — recreating what the pipeline does, with explanations

# 1. RFM Frequency Score
# WHY: Raw order counts don't account for how long someone has been a customer.
# A customer with 3 orders in 1 month is very different from 3 orders in 2 years.
# Binning into quintiles normalizes this and gives a relative ranking.
df['rfm_frequency_score'] = pd.qcut(
    df['total_orders'].rank(method='first'), q=5, labels=[1,2,3,4,5]
).astype(int)

# 2. RFM Monetary Score
# WHY: High-value customers who churn represent more revenue risk.
# Binning revenue into quintiles lets the model weight by value tier.
df['rfm_monetary_score'] = pd.qcut(
    df['total_revenue'].rank(method='first'), q=5, labels=[1,2,3,4,5]
).astype(int)

# 3. One-Time Buyer Flag
# WHY: EDA showed 74-76% of churned customers placed exactly 1 order.
# This is such a strong signal it deserves its own binary feature.
df['one_time_buyer'] = (df['total_orders'] == 1).astype(int)

# 4. High Return Flag
# WHY: Return rate > 50% suggests product dissatisfaction — a leading indicator of churn
# that isn't captured by purchase frequency alone.
df['high_return_flag'] = (df['return_rate'] > 0.5).astype(int)

# 5. Cancel Rate
# WHY: Cancellations signal intent to disengage. Different from returns
# (returns = dissatisfied post-purchase, cancellations = changed mind pre-fulfillment).
df['cancel_rate'] = df['cancelled_orders'] / (df['total_orders'] + 1)

# 6. Revenue Per Day
# WHY: Normalizes revenue by tenure. A customer who spent $500 in their first week
# is very different from one who spent $500 over 2 years.
df['revenue_per_day'] = df['total_revenue'] / (df['customer_tenure_days'] + 1)

# 7. Events Per Order
# WHY: Browsing-to-buying ratio. High events/order = lots of browsing, little buying.
# Low ratio = efficient, intentional purchaser (lower churn risk).
df['events_per_order'] = df['total_events'] / (df['total_orders'] + 1) if 'total_events' in df.columns else 0

print('=== Engineered Features Summary ===')
eng_features = ['rfm_frequency_score', 'rfm_monetary_score', 'one_time_buyer',
                'high_return_flag', 'cancel_rate', 'revenue_per_day']
eng_features = [f for f in eng_features if f in df.columns]
print(df[eng_features].describe().round(3).to_string())
```

```python
# Visualize engineered features vs churn
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Engineered Features vs Churn Rate', fontsize=13, fontweight='bold')

# RFM frequency score vs churn
rfm_churn = df.groupby('rfm_frequency_score')['churned'].mean()
axes[0].bar(rfm_churn.index, rfm_churn.values, color='#3498DB', alpha=0.8)
axes[0].set_title('Churn Rate by RFM Frequency Score')
axes[0].set_xlabel('Score (1=lowest, 5=highest frequency)')
axes[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
axes[0].set_xticks([1,2,3,4,5])

# Revenue per day vs churn
df['rev_per_day_bucket'] = pd.qcut(df['revenue_per_day'], q=5,
                                    labels=['Q1','Q2','Q3','Q4','Q5'])
rpd_churn = df.groupby('rev_per_day_bucket')['churned'].mean()
axes[1].bar(rpd_churn.index, rpd_churn.values, color='#E67E22', alpha=0.8)
axes[1].set_title('Churn Rate by Revenue/Day Quintile')
axes[1].set_xlabel('Quintile (Q1=lowest, Q5=highest)')
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

# One-time buyer vs churn
otb_churn = df.groupby('one_time_buyer')['churned'].mean()
axes[2].bar(['Repeat Buyer\n(2+ orders)', 'One-Time Buyer\n(1 order)'],
            otb_churn.values, color=[GREEN, RED], alpha=0.8)
axes[2].set_title('Churn Rate: One-Time vs Repeat Buyer')
axes[2].yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
for i, v in enumerate(otb_churn.values):
    axes[2].text(i, v + 0.01, f'{v:.1%}', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('eda_engineered_features.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Section 4 — Model Selection: Why XGBoost?

Three models were trained and compared. Here's the reasoning and the results.

```python
# Model selection rationale (printed, not just commented)
rationale = {
    'Logistic Regression': {
        'why_included': 'Baseline — interpretable, fast, good for linear relationships',
        'weakness': 'Assumes linear decision boundary; misses interaction effects between features',
        'expected': 'Decent AUC, lower precision on minority class'
    },
    'Random Forest': {
        'why_included': 'Handles non-linearity, robust to outliers, built-in feature importance',
        'weakness': 'Slower, can overfit on imbalanced data without class weighting',
        'expected': 'Strong AUC, good feature importance signal'
    },
    'XGBoost': {
        'why_included': 'Industry standard for tabular data, handles imbalance via scale_pos_weight, fast',
        'weakness': 'More hyperparameters to tune, less interpretable than Logistic Regression',
        'expected': 'Best AUC, sharpest feature importance'
    }
}

for model, info in rationale.items():
    print(f"\n{'='*50}")
    print(f"Model: {model}")
    for k, v in info.items():
        print(f"  {k.replace('_',' ').title()}: {v}")
```

```python
# Prepare clean feature set (no leakage — see Section 5)
LEAKAGE_COLS = [
    'days_since_last_order', 'rfm_recency_score', 'rfm_total',
    'purchase_events', 'total_events', 'approx_sessions',
    'revenue_per_session', 'cart_to_purchase_ratio',
]
META_COLS = ['user_id', 'city', 'state', 'top_category',
             'is_us_customer', 'churned', 'rev_per_day_bucket']
DROP_COLS  = LEAKAGE_COLS + META_COLS

feature_cols = [c for c in df.columns if c not in DROP_COLS]
df_model = df[feature_cols].copy()

for col in df_model.select_dtypes(include='object').columns:
    df_model[col] = LabelEncoder().fit_transform(df_model[col].astype(str))

X = df_model
y = df['churned']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
print(f'Feature set: {len(feature_cols)} features')
print(f'Train: {X_train.shape[0]:,} | Test: {X_test.shape[0]:,}')
print(f'Class imbalance ratio: {pos_weight:.2f} (handled via scale_pos_weight / class_weight)')
```

```python
# Train all 3 models and compare
models = {
    'Logistic Regression': LogisticRegression(
        max_iter=1000, class_weight='balanced', random_state=RANDOM_STATE
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=200, class_weight='balanced',
        max_depth=8, random_state=RANDOM_STATE
    ),
    'XGBoost': XGBClassifier(
        n_estimators=300, max_depth=5, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        scale_pos_weight=pos_weight,
        use_label_encoder=False, eval_metric='logloss',
        random_state=RANDOM_STATE
    )
}

results = {}
for name, model in models.items():
    Xtr = X_train_s if name == 'Logistic Regression' else X_train
    Xte = X_test_s  if name == 'Logistic Regression' else X_test
    model.fit(Xtr, y_train)
    probs = model.predict_proba(Xte)[:, 1]
    preds = model.predict(Xte)
    auc   = roc_auc_score(y_test, probs)
    results[name] = {'model': model, 'probs': probs, 'preds': preds, 'auc': auc}
    print(f'{name}: ROC-AUC = {auc:.4f}')
```

```python
# ROC curves + feature importance side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Model Comparison', fontsize=13, fontweight='bold')

colors = {'Logistic Regression': '#3498DB', 'Random Forest': '#2ECC71', 'XGBoost': '#E67E22'}
for name, r in results.items():
    RocCurveDisplay.from_predictions(
        y_test, r['probs'],
        name=f"{name} (AUC={r['auc']:.3f})",
        ax=axes[0], color=colors[name]
    )
axes[0].set_title('ROC Curves — All Models')
axes[0].plot([0,1],[0,1],'k--',alpha=0.3)

importances = pd.Series(
    results['XGBoost']['model'].feature_importances_, index=X.columns
).sort_values(ascending=False).head(12)
importances.plot(kind='barh', ax=axes[1], color='#E67E22', alpha=0.85)
axes[1].set_title('Feature Importances (XGBoost)')
axes[1].invert_yaxis()
axes[1].set_xlabel('Importance Score')

plt.tight_layout()
plt.savefig('model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

best = max(results, key=lambda k: results[k]['auc'])
print(f'\nWinner: {best} (AUC: {results[best]["auc"]:.4f})')
print(f'Top 3 features: {list(importances.head(3).index)}')
```

## Section 5 — Leakage Detection: The Most Important Decision

This section documents the single most important analytical decision in the project — identifying and removing data leakage that inflated model performance to a meaningless 1.0 AUC.

### What is data leakage?
Leakage occurs when a feature used in training encodes information about the target variable that would not be available at prediction time. In simple terms: the model "cheats" by using the answer to predict the answer.

### What we found
The first model run returned **ROC-AUC = 1.0 across all three models**. This is almost never a good sign.

```python
# Demonstrate the leakage problem
leakage_features = [
    ('days_since_last_order',
     'IS the churn definition (churned = no order in 90+ days). Using this is just memorizing the label.'),
    ('rfm_recency_score',
     'Binned version of days_since_last_order. Same leakage, just disguised.'),
    ('rfm_total',
     'Includes recency score. Carries the leakage through combination.'),
    ('purchase_events',
     'Event table logs every purchase — mirrors order count 1:1. Indirect leakage.'),
    ('total_events',
     'Dominated by purchase_events. Carries the indirect leakage.'),
    ('approx_sessions',
     'Derived from event timestamps — encodes when someone last was active.'),
    ('revenue_per_session',
     'Derived from total_events. Same leakage chain.'),
    ('cart_to_purchase_ratio',
     'Derived from purchase_events. Same leakage chain.'),
]

print('=== Identified Leaky Features ===')
print(f'{"Feature":<30} {"Why it leaks"}')
print('-' * 80)
for feat, reason in leakage_features:
    print(f'{feat:<30} {reason}')

print('\n=== Key Insight ===')
print('The structural issue: churn is defined as "no purchase in 90 days".')
print('Any feature derived from purchase timing or event activity')
print('mathematically encodes the churn label. Removing them is non-negotiable.')
```

```python
# Show the leak detector output — correlation check
df_temp = df[feature_cols].copy()
for col in df_temp.select_dtypes(include='object').columns:
    df_temp[col] = LabelEncoder().fit_transform(df_temp[col].astype(str))

corr_with_churn = df_temp.join(y).corr()['churned'].abs().sort_values(ascending=False)
corr_with_churn = corr_with_churn.drop('churned')

plt.figure(figsize=(10, 6))
colors_bar = ['#E74C3C' if v > 0.6 else '#3498DB' for v in corr_with_churn.values]
plt.barh(corr_with_churn.index, corr_with_churn.values, color=colors_bar, alpha=0.85)
plt.axvline(x=0.6, color='red', linestyle='--', alpha=0.7, label='Leakage threshold (0.6)')
plt.xlabel('Absolute Correlation with Churn Label')
plt.title('Leak Detector: Feature Correlation with Target\n(red = potential leakage)', fontweight='bold')
plt.legend()
plt.tight_layout()
plt.savefig('leakage_detection.png', dpi=150, bbox_inches='tight')
plt.show()

high_corr = corr_with_churn[corr_with_churn > 0.6]
if len(high_corr) > 0:
    print(f'\n⚠️  High correlation features (potential leakage): {list(high_corr.index)}')
else:
    print('\n✅ No features with correlation > 0.6 — clean feature set')
```

```python
# Reframe: scoring engine vs forecasting model
print('=== Project Reframe: Scoring Engine vs Forecasting Model ===')
print()
print('After removing leaky features, AUC settled at ~0.98-0.99.')
print('This is still high — because the remaining features (frequency score,')
print('revenue per day, one-time buyer flag) are genuinely strong predictors.')
print()
print('The model is not "cheating" — it is doing exactly what a risk scoring')
print('engine should do: cleanly separating customers by behavioral risk tier.')
print()
print('This is how production tools like Gainsight and Totango work:')
print('they score customers on observable signals, not future outcomes.')
print()
print('Interview answer: "I identified temporal leakage inherent to retrospective')
print('churn labeling, removed the offending features, and reframed the project')
print('as a behavioral risk scoring engine — which is how churn tools actually')
print('work in production. The high AUC reflects strong signal separation, not leakage."')
```

```python
# Final summary visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('The Core Finding That Shaped Everything', fontsize=13, fontweight='bold')

# Left: orders per month comparison
retained_opm = df[df['churned']==0]['orders_per_month'].clip(upper=5)
churned_opm  = df[df['churned']==1]['orders_per_month'].clip(upper=5)
axes[0].hist(retained_opm, bins=30, alpha=0.7, label=f'Retained (avg={retained_opm.mean():.2f}/mo)', color=GREEN)
axes[0].hist(churned_opm,  bins=30, alpha=0.7, label=f'Churned (avg={churned_opm.mean():.2f}/mo)',  color=RED)
axes[0].set_title('Orders per Month: The Key Differentiator')
axes[0].set_xlabel('Orders per Month')
axes[0].legend()

# Right: the project flow
axes[1].axis('off')
flow = [
    ('BigQuery SQL', 'Join 4 tables, define churn,\nengineer raw features'),
    ('EDA', 'Find: purchase frequency is\nthe core separator'),
    ('Feature Engineering', 'Build RFM scores, one-time\nbuyer flag, revenue/day'),
    ('XGBoost', 'Train scorer, detect leakage,\nreframe as scoring engine'),
    ('Tableau + Claude AI', 'Surface $6.3M revenue at risk\nwith actionable retention briefs'),
]
y_pos = 0.92
for step, desc in flow:
    axes[1].text(0.05, y_pos, f'→ {step}', fontsize=11, fontweight='bold',
                 transform=axes[1].transAxes, color='#2C3E50')
    axes[1].text(0.05, y_pos-0.06, f'   {desc}', fontsize=9,
                 transform=axes[1].transAxes, color='#7F8C8D')
    y_pos -= 0.18
axes[1].set_title('ChurnLens: The Full Pipeline')

plt.tight_layout()
plt.savefig('project_summary.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Summary

### What this analysis revealed

1. **This is an activation problem, not a retention problem.** ~75% of high-risk customers never made a second purchase. The business needs to fix the post-first-purchase experience, not just run win-back campaigns.

2. **Purchase frequency is everything.** Retained customers order 10-15x more frequently than churned ones. This single insight shaped the entire model, dashboard, and chatbot.

3. **Email acquisition dominates both risk and revenue.** The largest at-risk segments are all email-acquired — but so are the most retained customers. Email quality, not quantity, is what differentiates.

4. **Facebook × 45-54 is the highest-risk acquisition segment.** Paid social targeting this demographic produces customers with the lowest repeat purchase rates.

5. **Leakage detection is as important as model performance.** An AUC of 1.0 is a red flag, not a win. The analytical decision to identify and remove leaky features — and reframe the project as a scoring engine — is the most technically mature part of this project.
